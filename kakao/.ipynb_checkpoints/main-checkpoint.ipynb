```python
import numpy as np
import pandas as pd
import seaborn as sns
from datetime import datetime, timedelta
import re
import matplotlib as mpl
import matplotlib.pyplot as plt
font_name = mpl.font_manager.FontProperties(fname='C:/Windows/Fonts/malgun.ttf').get_name()
mpl.rc('font', family=font_name)
```

```python
# 변수 선언 및 정규화
text = []
text2 = []
texttime = []
textname = []

user = ['김연수', '백민혁', '주용', '허대현']
time1 = datetime(2020, 5, 27)
time2 = datetime.now()
day = ((time2-time1).days)
pic = re.compile('\[.*\] \[오.*\] 사진\n')
pp = []
dele = re.compile('\[.*\] \[오.*\] 삭제된 메시지입니다.\n')
dd = []
emo = re.compile('\[.*\] \[오.*\] 이모티콘\n')
ee = []

with open("data/KakaoTalk.txt", "r", encoding="utf-8") as f:
    lines = f.readlines()
    for line in lines[5:]:
        if '] [' in line:
            pp.extend(pic.findall(line))
            dd.extend(dele.findall(line))
            ee.extend(emo.findall(line))
            name = line.split('] ')[0][1:]
            text.append(name+"*-+/"+line.split('] ')[2].replace('사진\n','').replace('이모티콘\n','').replace('삭제된 메시지입니다',''))
            texttime.append(line.split('] ')[1][1:].replace('사진\n','').replace('이모티콘\n','').replace('삭제된 메시지입니다',''))
            textname.append(name)

for tar in text:
    text2.append(tar.replace('\n',''))
```

```python
# 가장 긴 단어
# 링크빼고

manyw = ""
manyww = []
mass = []
mass1 = []
mass2 = []
for tar2 in text2:
    namet = tar2.split('*-+/')
    manyw += namet[1]+' '

manyww = manyw.replace("  ", " ").split(' ')
for i in range(0,11):
    maany = [0,""]
    for i in range(len(manyww)):
        if maany[0] < len(manyww[i]):
            maany[0] = len(manyww[i])
            maany[1] = manyww[i]
    mass.append(maany)
    manyww.remove(maany[1])

#print(mass)
for M in range(len(mass)):
    mass1.append(mass[M][0])
    mass2.append(mass[M][1])
for M in range(len(mass2)):
    mass2[M] = mass2[M][0:8]
    if mass2[M] == "https://":
        mass2[M] += "%s" %M

#print(mass) #어떤 메시지인지 보고싶으면 해제
fig, ax = plt.subplots(facecolor="w", figsize=(15,10))
plt.bar(mass2, mass1, width=0.6)
plt.title("제일 긴 메시지", fontsize = 25)
for i, v in enumerate(mass2):
    plt.text(v, mass1[i], str(mass1[i])+' 자', fontsize = 15,  color='blue', horizontalalignment='center', verticalalignment='bottom')
plt.xlabel('메시지')
plt.ylabel('글자수')
plt.show()
```

```python
# 총 대화 횟수

namecount = [0,0,0,0]
for tar2 in text2:
    namet = tar2.split('*-+/')
    for i in range(len(user)):
        if namet[0]==user[i]:
            namecount[i] += 1
            
#print(namecount)

fig, ax = plt.subplots(facecolor="w", figsize=(15,10))
plt.bar(user, namecount, width=0.6)
for i, v in enumerate(user):
    plt.text(v, namecount[i], str(namecount[i])+' 번', fontsize = 15,  color='blue', horizontalalignment='center', verticalalignment='bottom')
plt.title("유저별 톡 수", fontsize = 25)
plt.xlabel('이름')
plt.ylabel('보낸 메시지 수')
plt.show()
```

```python
# 삭제 많이 한 놈

delcount = [0,0,0,0]
ddel = []
for i in range(len(dd)):
    dd[i] = dd[i].replace('\n', '').replace('[', '').replace(']', '')
    ddel.append(dd[i].split(' ')[0]+'abcd'+dd[i].split(' ')[3])
    ddel[i] = ddel[i].split('abcd')[0]
    for j in range(len(user)):
        if ddel[i]==user[j]:
            delcount[j] += 1

#print(delcount) #삭제한 횟수
fig, ax = plt.subplots(facecolor="w", figsize=(15,10))
plt.bar(user, delcount, width=0.6)
for i, v in enumerate(user):
    plt.text(v, delcount[i], str(delcount[i])+' 번', fontsize = 15,  color='blue', horizontalalignment='center', verticalalignment='bottom')
plt.title("메시지 삭제한 횟수", fontsize = 25)
plt.xlabel('이름')
plt.ylabel('보낸 메시지 수')
plt.show()
```

```python
# 특정 단어 언급 횟수(초성제외)

target = ["친구"]

wordcount = [0,0,0,0]
wordmemory = []

for tar2 in text2:
    namet = tar2.split('*-+/')
    for i in range(len(user)):
        if namet[0]==user[i]:
            for w in target:
                if namet[1].count(w)>0:
                    wordcount[i] += tar2.count(w)
                    wordmemory.append(namet[0]+": "+namet[1])
            
#print(wordcount)
#print(wordmemory) #뭔 말 했는지 알고 싶으면 주석 해제
fig, ax = plt.subplots(facecolor="w", figsize=(15,10))
plt.bar(user, wordcount, width=0.6)
for i, v in enumerate(user):
    plt.text(v, wordcount[i], str(wordcount[i])+' 번', fontsize = 15,  color='blue', horizontalalignment='center', verticalalignment='bottom')
plt.title("%s을/를 외친 횟수" % target[0], fontsize = 25)
plt.xlabel('이름')
plt.ylabel('보낸 메시지 수')
plt.show()
```

```python
#시간대별 전체 톡
#6분마다 취합

hourcount=[]
hour=[]
for i in range(1, 24*10+1):
    hourcount.append(str(i)+"시")
    hour.append(0)
for tar2 in texttime:
    time = tar2.split(' ')
    timer = time[1].split(':')
    if time[0] == '오후':
        timer[0] = str(int(timer[0])+12)
    timer[1] = np.around(int(timer[1])/360*100)
    hour[(int(timer[0])-2)*10+int(timer[1])] += 1
for i in range(len(hour)):
    hour[i] = int(hour[i])/day

fig, ax = plt.subplots(facecolor="w", figsize=(15,10))
plt.plot(np.arange(0,24,0.1), hour)
plt.title("하루 시간대별 전체 톡 빈도", fontsize = 25)
plt.xlabel('시간')
plt.ylabel('보낸 메시지 수')
plt.show()
```
